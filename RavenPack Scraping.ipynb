```python
'''
Scrape RavenPack

returns JSON with entries: {ts = time stamp, and requested variable}
'''
import json, urllib
def get_ravenpack(country, variable):
    url = 'https://coronavirus.ravenpack.com/data/1.3/country/'+country+'/'+variable+'.json'
    page = urllib.request.urlopen(url)
    data = json.loads(page.read())
    return data
    
'''
Example

'''
get_ravenpack('US','panic')
```
